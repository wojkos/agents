## Krok 3: Tworzenie Agenta
Definiujemy osobowo≈õƒá naszego Agenta.

```python
from adk import Agent

def create_my_agent():
    agent = Agent(
        name="SzkolnyPomocnik",
        instructions="""
        Jeste≈õ pomocnikiem ucznia.
        Odpowiadasz z poczuciem humoru, ale dbasz o poprawno≈õƒá merytorycznƒÖ.
        U≈ºywasz czystego tekstu, unikasz ikon.
        """
    )
    return agent

my_agent = create_my_agent()
print("‚úÖ Agent stworzony!")
```
